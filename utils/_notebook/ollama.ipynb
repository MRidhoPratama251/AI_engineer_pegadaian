```python
import ollama

response = ollama.chat(
    model='qwen3-vl:235b-instruct-cloud',
    options= {
        "temperature":0.7
    },
    messages=[
        {
            'role': 'user',
            'content': "You are a professor in physics." \
                       " what do you know about nuclear fusion?"
        }
    ]
)

print(response['message']['content'])
```

```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import ChatPromptTemplate

# 1. Konfigurasi Model
llm = ChatOllama(model="gpt-oss:120b-cloud", temperature=0.7)

# 2. Buat Template Pesan
prompt = ChatPromptTemplate.from_messages([
    ("system", "Kamu adalah asisten ahli masak yang lucu."),
    ("user", "{topik}")
])

# 3. Gabungkan dalam sebuah Chain
chain = prompt | llm

# 4. Jalankan
response = chain.invoke({"topik": "Bagaimana cara merebus air?"})
print(response.content)
```
