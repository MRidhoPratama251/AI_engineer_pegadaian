```python
import hashlib
import base64

def generate_short_id(input_string: str, salt: str = "my_secret_salt"):
    # 1. Tambahkan salt agar hasil unik & tidak mudah ditebak
    full_string = input_string + salt
    
    # 2. Gunakan SHA-256 untuk hashing
    hash_obj = hashlib.sha256(full_string.encode())
    
    # 3. Encode ke Base64 agar menjadi alfanumerik pendek
    # Kita gunakan urlsafe_b64encode agar tidak ada karakter '/' atau '+'
    short_id = base64.urlsafe_b64encode(hash_obj.digest()).decode()
    
    # 4. Ambil 6 karakter pertama
    return short_id[:6]

# Contoh penggunaan
print(generate_short_id("RIDHO-123!@#")) # Output: e.g. "x7A2nQ"
print(generate_short_id("ORDER-2026-XYZ")) # Output: e.g. "kP9m1B"
```
